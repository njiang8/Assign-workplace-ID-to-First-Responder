# Get Fire Station wrkID
#Firstly, load the joined popu data

```python
import pandas as pd
import numpy as np
```

```python
#Get Data that is population in each wrkplace
sumpop = pd.read_csv('/Users/jiangna/Desktop/distfirepopsum.csv')
```

```python
#Remove a useless column in sumpop
sumpop = sumpop.drop(['Unnamed: 0'],axis = 1)
```

```python
sumpop = sumpop.drop(['wrkID.1'],axis = 1)
```

```python
sumpop.head()
```

```python
#Get Table which sort by the distance to each pol sta
wpop = pd.DataFrame()
for a in range(2,215):
    df1 = sumpop.iloc[:,[0,1,a]]
    b = str(a-2)
    df1 = df1.sort_values(by = b).reset_index().drop(['index'],axis =1).drop(b,axis = 1)
    df1['HPF'] = 3001 + a-2
    wpop = wpop.append(df1)
```

```python
wpop.head()
```

```python
wpop.HPF.unique()
```

# Groupby unique value which represent each police station

```python
#Get groups based on Type
tgroup = wpop.groupby(wpop['HPF'])
```

```python
#New DataFrame to hold the results
wp = pd.DataFrame()
#Nested For Loops to Seltect the Group Fisrt and Then To the Selection by population Sum
for i in range (3001,3214):
    tdf = tgroup.get_group(i)
    psum = 0
    for j in range (0,len(tdf)):
        #Use while to Set the Calculation bondary by thing the hospital population
        while psum < 50:
            psum = psum + tdf.iloc[j,1]
            wdf = tdf.iloc[0:j,0]
            #print(bond[i-1])
            break
    wp = wp.append(wdf,ignore_index = True)
    wpt = wp.T
```

```python
wpt
```

```python
#convert Multi Col into One Col
wrkpl = pd.melt(wpt).dropna(subset=['value'])
```

```python
print(wrkpl.head())
```

```python
wrkpl.columns = ['HPF', 'wrkId']
```

```python
wrkpl['HPF1'] = wrkpl.iloc[:,0] + 3001
```

```python
wrkpl = wrkpl.drop(['HPF'],axis = 1)
```

```python
wrkpl = wrkpl.rename(columns={'HPF1': 'HPF'})
```

```python
wrkpl.head()
```

```python
wrkpl.to_csv('/Users/jiangna/Desktop/fire_wrkid.csv')
```
