```python
import pandas as pd
import numpy as np
```

# First load the population and drop useless columns

```python
df = pd.read_csv('/Users/jiangna/Desktop/GMU_CSS_Work/A1_GCHW/O_Sheets/population_final.csv')
```

```python
df1 = df.dropna(subset=['wp'])
df2 = df1.drop(['id', 'age','hhold_type','lat','long','school','sex','hhold','tract'], axis=1).reset_index().drop(['index'], axis =1)
df2.head()
```

```python
df2['count'] = 1
```

```python
df2.head()
```

```python
#Get count of teach wrkID which is the population in each wrkId
workpop = df2.groupby(['wp'], as_index=False).count()
```

```python
workpop.head()
```

```python
workpop = workpop.rename(columns={'wp': 'wrkID'})
workpop.head()
```

# Secondly, load the distance result

```python
wrkdist = pd.read_csv('/Users/jiangna/Desktop/poldist_f.csv')
```

```python
wrkdist.head()
```

```python
#join two data frame based on "wrkID"
poldistpop = workpop.merge(wrkdist, on="wrkID", how = 'inner')
```
